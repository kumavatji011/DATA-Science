```python
import numpy as np
```

```python
# ============================================
# 1. Create a 1D NumPy array from 1 to 20
# ============================================
```

```python
arr = np.arange(1, 21)
print("1D Array:", arr)

# a. Calculate sum, mean, median, and standard deviation
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))

# b. Find indices of elements greater than 10
indices = np.where(arr > 10)
print("Indices of elements greater than 10:", indices)
```

```python
# ============================================
# 2. Create a 2D NumPy array of shape 4x4
# ============================================
```

```python
array_2d = np.arange(1, 17).reshape(4, 4)
print("\n2D Array:\n", array_2d)

# a. Transpose of the array
transpose_array = array_2d.T
print("\nTranspose of the array:\n", transpose_array)

# b. Row-wise and column-wise sums
row_sum = np.sum(array_2d, axis=1)
column_sum = np.sum(array_2d, axis=0)

print("\nRow-wise sum:", row_sum)
print("Column-wise sum:", column_sum)
```

```python
# ============================================
# 3. Create two random 3x3 arrays
# ============================================
```

```python
arr1 = np.random.randint(1,21,(3 ,3))
arr2 = np.random.randint(1,21,(3,3))

print('\nFirst Random Array:\n',arr1)
print('\nSecond Random Array:\n',arr2)

#a . Element wise operations
print("\n Element-wise Additions:\n",arr1+arr2)
print("\n Element-wise Substractons:\n",arr1-arr2)
print("\n Element-wise Multiplications:\n",arr1*arr2)

#b. Dot product of the arrays
dot_product = np.dot(arr1,arr2)
print("\nDot Product:\n",dot_product)
```

```python
# ============================================
# 4. Reshape and slice an array
# ============================================
```

```python
#Create a 1D array of size 12
arr = np.arange(1,13)

#Reshape into 3 x 4 array
reshaped_array = arr.reshape(3,4)
print("\nReshaped 3x4 Array : \n",reshaped_array)

#Slice first two rows and last two columns 
sliced_array = reshaped_array[:2 ,-2]
print("\n Sliced Array (frist two rows , last two columns):\n",sliced_array)
```
