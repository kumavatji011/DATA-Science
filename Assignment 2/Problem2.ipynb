```python
#import pandas

import pandas as pd
```

```python
# 1.Create the DataFrame 
data={
    'Name':['Alice','Bob','charlie','Diana','Eve'],
    'Age':[27,32,44,53,21],
    'Salary': [40000,23000,22222,44444,12345],
    'Department':['HR','Finance','IT','Marketing','HR']
}
df= pd.DataFrame(data)
```

```python
#a. Print the first five rows 
print("First five rows of the Dataframe:\n")
print(df.head())
```

```python
#b. Summary Statistics for age and salary
print("\n Summary statistics for age and Salary:")
print(df[['Age','Salary']].describe())
```

```python
#c. Average salary of employess in the HR department
avg_hr_salary=df[df['Department']=='HR']['Salary'].mean
print("\n Average salary in HR department:",avg_hr_salary)
```

```python
#2. Filter employees aged between 25 and 30
filtered_df =df [(df['Age']>=25)&(df['Age']<=30)]
print("\nEmplyolees aged between 25 and 30:")
print(filtered_df)
```

```python
#3. Add a new column 'Bonus' (10% of Salary)
df['Bonus']=df['Salary']*0.10
print("\nDataFrame with Bonus column:")
print(df)
```

```python
#4. Group by Depaartment and Calculate average salary 
avg_salary_by_dept =df.groupby('Department')['Salary'].mean()
print("\nAverage salary by department:")
print(avg_salary_by_dept)
```

```python
#5. Sort by salary in ascending order and save to CSV 
sorted_df=df.sort_values(by="Salary",ascending=True)
sorted_df.to_csv("sorted_employees.csv",index=False)

print("\n Sorted Dataframe saved to 'sorted_employees.csv'")
```
