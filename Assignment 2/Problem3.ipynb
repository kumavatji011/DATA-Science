```python
#import required libraries 
import matplotlib.pyplot as plt
import random
```

```python
# -------------------------------
# 1. Line Plot
# -------------------------------
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.figure()
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()
```

```python
# -------------------------------
# 2. Bar Graph
# -------------------------------
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.figure()
plt.bar(students, marks, color=['blue', 'green', 'orange', 'red'])
plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
# -------------------------------
# 3. Pie Chart
# -------------------------------
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = (0.1, 0, 0, 0)  # Highlight highest revenue region

plt.figure()
plt.pie(revenue, labels=regions, autopct='%1.1f%%',
        explode=explode, startangle=140)
plt.title("Revenue Distribution by Region")
plt.show()
```

```python
# -------------------------------
# 4. Histogram
# -------------------------------
```

```python
random_numbers = [random.randint(1, 100) for _ in range(1000)]

plt.figure()
plt.hist(random_numbers, bins=10, color='purple', edgecolor='black')
plt.title("Histogram of Random Numbers (1100)")
plt.xlabel("Number Range")
plt.ylabel("Frequency")
plt.show()
```
